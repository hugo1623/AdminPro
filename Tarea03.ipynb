```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('top10s.csv', encoding='ISO-8859-1')
```

# 1.- Listar solo los géneros 

```python
data.loc[:, ("top genre")]
```

# Filtrar Canciones de Longitud mayor que el promedio

```python
data
```

#.  Mostrar solo la canción el artista y la duración de la canción. 

```python
data.loc[:, ("title", "artist", "dur")]
```

# Listar los artistas que hicieron canciones en el 2019

```python
data.loc[data['year'] >= 2019, ['artist','year'] ]
```

#.   Mostrar solo la columna del artista y su canción

```python
data.groupby(["artist", "title"]).sum()
```

# Top de canciones más ruidosas (dB)

```python
Canciones_ruidosas = data.groupby([ "title","dB"]).count()
top_CancionesRuidosas = Canciones_ruidosas.sort_values(by = "dB", ascending= False).head()
top_CancionesRuidosas
```

# Top de canciones más largas

```python
Canciones_MasLargas = data.groupby([ "title","dur"]).count()
top_CancionesMasLargas = Canciones_MasLargas.sort_values(by = "dur", ascending= False).head()
top_CancionesMasLargas
```

# Top de canciones más ruidosas y más largas.

```python
Canciones_MasRuidosaLargas = data.groupby([ "title","dB","dur"]).count()
Canciones_MasRuidosaLargas
```

```python
top_MasRuidosaLargas = Canciones_MasRuidosaLargas.sort_values(by = ["dB","dur"], ascending= False).head()
top_MasRuidosaLargas
```

# Agrupar y graficar:

# Generos con las canciones más largas de duración más largos  (agrupar por género y sacar máximo de canción)



```python
Canciones_MasLargas = data.groupby([ "top genre","title","dur"]).count()
top_CancionesMasLargas = Canciones_MasLargas.sort_values(by = "dur", ascending= False).head()
top_CancionesMasLargas
```

```python
top_CancionesMasLargas.reset_index(inplace=True)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(top_CancionesMasLargas.dur, top_CancionesMasLargas.title)
plt.show()
```

# Años con su promedio de bailable (agrupar por año y sacar el promedio de dnce)



```python

años_PromedioBaile = data.groupby([ "title","year","dnce"]).count()
Anos_Promedio = años_PromedioBaile.sort_values(by = "dnce", ascending= False).head()
Anos_Promedio

```

```python
Anos_Promedio.reset_index(inplace=True)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(Anos_Promedio.year, Anos_Promedio.dnce)
plt.show()
```

# Años con su promedio de ruido (dB)

```python
años_PromedioRuido = data.groupby([ "title","year","dB"]).count()
Anos_PromedioRuido = años_PromedioRuido.sort_values(by = "dB", ascending= False).head()
Anos_PromedioRuido
```

```python
Anos_PromedioRuido.reset_index(inplace=True)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(Anos_PromedioRuido.year, Anos_PromedioRuido.dnce)
plt.show()
```
